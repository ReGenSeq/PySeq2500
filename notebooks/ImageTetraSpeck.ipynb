```python
import logging

LOGLEVEL = logging.DEBUG
# LOGLEVEL = logging.INFO
```

```python
image_area = {"flowcell": "A", "LLx": 12, "LLy": 35, "URx": 11, "URy": 34, "overlap": 0}
focus = {
    "routine": "full once",
    "optics": {
        "power": {"green": 100, "red": 100},
        "filter": {"green": "open", "red": "open"},
    },
}
image = {
    "optics": {
        "power": {"green": 200, "red": 200},
        "filter": {"green": "open", "red": "open"},
    },
    "nz": 3,
}
```

```python
from pyseq2500.sequencer import PySeq2500
from pyseq2500.microscope import Microscope
from pyseq_core.base_protocol import BaseROI
import sys

LOGGER = logging.getLogger("PySeq")
console_handler = logging.StreamHandler(sys.stdout)
console_handler.setFormatter(
    logging.Formatter("%(asctime)s - %(levelname)s - %(message)s")
)
console_handler.setLevel(LOGLEVEL)
LOGGER.addHandler(console_handler)
LOGGER.setLevel(LOGLEVEL)
```

```python
stage = PySeq2500.custom_roi_stage(
    **image_area, stage={"tilt1": 21800, "tilt2": 21800, "tilt3": 21800}
)
roi = BaseROI(name="tetraspeck", stage=stage, focus=focus, image=image)
```

```python
microscope = Microscope()
```

```python
await microscope._connect()
await microscope._configure()
await microscope._initialize()
```

```python
await microscope._tilt_stack(roi, tilt_init=21000, tilt_last=23000, tilt_step=100)
```

```python
await microscope._find_focus(roi)
```

```python
await microscope._set_parameters(roi.image.optics)
await microscope._scan(roi)
```

```python
await microscope._shutdown()
```
